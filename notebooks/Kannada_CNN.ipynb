```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns

#############################################
# 1. TRAINING: Load Data, Build, Train Model
#############################################

# Parameters
data_dir = "/content/Kannada_Alphabets"  # Update this path as needed
img_height = 28
img_width = 28
batch_size = 32
seed = 123

# Create training and validation datasets using validation_split.
# Here we use label_mode="categorical" so that targets are one-hot encoded.
train_dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=seed,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    color_mode="grayscale",
    label_mode="categorical"  # one-hot encoding
)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=seed,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    color_mode="grayscale",
    label_mode="categorical"  # one-hot encoding
)

# Get class names (both datasets have the same class names)
class_names = train_dataset.class_names
print("Classes:", class_names)

# Apply data augmentation to the training dataset only
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomTranslation(0.1, 0.1)
])
train_dataset = train_dataset.map(lambda x, y: (data_augmentation(x, training=True), y))

# Build the CNN Model
model = tf.keras.models.Sequential([
    # Block 1
    tf.keras.layers.Conv2D(32, (3, 3), padding="same", activation="relu",
                           input_shape=(img_height, img_width, 1)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    # Block 2
    tf.keras.layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    # Block 3
    tf.keras.layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),

    # Dense layers
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(len(class_names), activation="softmax")
])
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()

# Train the model with callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=5, restore_best_weights=True, verbose=1
)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6, verbose=1
)
epochs = 480
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs,
    callbacks=[early_stopping, reduce_lr]
)

# Save the full model (architecture + weights)
model_save_path = "full_model.h5"
model.save(model_save_path)
print("Full model saved to", model_save_path)

#############################################
# 2. INFERENCE & EVALUATION
#############################################

# Load the saved full model
loaded_model = tf.keras.models.load_model(model_save_path)
print("Loaded model from", model_save_path)

# Function to plot random images from the validation set with true and predicted labels
def plot_random_images_with_predictions(dataset, model, class_names, num_images=12):
    # Unbatch the dataset to get all examples
    data_list = list(dataset.unbatch().as_numpy_iterator())
    # Randomly select indices from the full dataset
    indices = np.random.choice(len(data_list), size=num_images, replace=False)

    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(indices):
        img, label = data_list[idx]
        # Expand dims for model prediction
        img_expanded = np.expand_dims(img, axis=0)
        prediction = model.predict(img_expanded)
        pred_label = class_names[np.argmax(prediction)]
        true_label = class_names[np.argmax(label)]

        plt.subplot(3, 4, i+1)
        plt.imshow(img.reshape(img_height, img_width), cmap="gray")
        plt.title(f"True: {true_label}\nPred: {pred_label}")
        plt.axis("off")
    plt.tight_layout()
    plt.show()

# Plot random images from the validation dataset with predictions
plot_random_images_with_predictions(val_dataset, loaded_model, class_names)

# Compute overall evaluation metrics on the validation set
all_true = []
all_pred = []
for images, labels in val_dataset:
    preds = loaded_model.predict(images)
    all_true.extend(np.argmax(labels.numpy(), axis=1))
    all_pred.extend(np.argmax(preds, axis=1))
cm = confusion_matrix(all_true, all_pred)
acc = accuracy_score(all_true, all_pred)
report = classification_report(all_true, all_pred, target_names=class_names)

print("Confusion Matrix:")
print(cm)
print("Accuracy: {:.2f}%".format(acc * 100))
print("Classification Report:")
print(report)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()




```

```python
# Function to classify a single JPG image
def predict_label(image_path, model, class_names, img_height=28, img_width=28):
    img = tf.keras.preprocessing.image.load_img(
        image_path,
        target_size=(img_height, img_width),
        color_mode="grayscale"
    )
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    prediction = model.predict(img_array)
    return class_names[np.argmax(prediction)]

def classify_image(image_path):
    label = predict_label(image_path, loaded_model, class_names)
    print("Predicted label:", label)
classify_image("/content/Kannada_Alphabets/14/0040.jpg")
```
